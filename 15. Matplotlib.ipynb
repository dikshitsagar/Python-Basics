```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2,3,5,6,7,0,2])
plt.ylabel('Numbers')
plt.xlabel('Indices')
plt.title('My Plot')
plt.show()
# if no x-axis values are given then 1,2,3.. becomes the indicies
# y-axis --> values
# x-axis --> indexed of values
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.xlabel('Numbers')
plt.ylabel('Squares')
plt.grid()
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16],'ro') # here, 'r' is red color, 'o' is shape of the data point
plt.xlabel('Numbers')
plt.ylabel('Squares')
plt.grid()
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16],'b*') # here, 'b' is the blue color, '*' is the shape of the data point.
plt.xlabel('Numbers')
plt.ylabel('Squares')
plt.grid()
plt.show()
```

```python
import numpy as np 
t = np.arange(0.,5.,0.2)

plt.plot(t,t**2,'b--',label = '^2') # blue colored dashed lines
plt.plot(t,t**2.2,'rs',label = '^2.2') # red colored squares
plt.plot(t,t**2.5,'g^',label = '^2.5') # green color triangle

plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16], linewidth = 5)
plt.xlabel('Numbers')
plt.ylabel('Squares')
plt.grid()
plt.show()
```

```python
x1 = [1,2,3,4]
y1 = [1,4,9,16]
x2 = [1,2,3,4]
y2 = [2,4,6,8]

lines = plt.plot(x1,y1,x2,y2)

plt.setp(lines[0], color = 'r', linewidth = 3)
plt.setp(lines[1], color = 'b', linewidth = 3)
plt.grid()
plt.show()
```

```python
plt.figure(1)

plt.subplot(211) # numrows, numcolumns,figno
plt.grid()
plt.plot(x1,y1,'b--')


plt.subplot(212) # numrows, numcolumns,figno
plt.grid()
plt.plot(x2,y2,'go-')
```

```python
# with figure(2)

plt.figure(1)

plt.subplot(211) # numrows, numcolumns,figno
plt.grid()
plt.plot(x1,y1,'b--')

plt.figure(2)
plt.subplot(211) # numrows, numcolumns,figno
plt.grid()
plt.plot(x2,y2,'go-')
plt.figure(2)
plt.subplot(212) # numrows, numcolumns,figno
plt.grid()
plt.plot(x2,y2,'ro-')
```
